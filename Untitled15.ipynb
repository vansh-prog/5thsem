```python
import numpy as np

# Create a dataset of continuous data
data = np.random.normal(size=100)
print(data)
# Discretize the data using equal-width bins
bins = np.linspace(min(data), max(data), 4)
print(bins)
discretized_data = np.digitize(data, bins)

# Print the discretized data
print(discretized_data)
```
