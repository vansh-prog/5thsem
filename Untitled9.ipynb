```python
#importing the data
import pandas as pd
df = pd.read_csv('dataset/auto_mpg.csv')
df.info()    
```

```python
#replacing the null values in 'horsepower' with its mean value
df['horsepower'].fillna(df['horsepower'].mean(), inplace=True)
df.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= df['horsepower']
y= df['mpg']
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
#plotting the lines
ax.scatter(x,y)
```

```python
df.groupby('model_year').count()['name']
```

```python
import matplotlib.pyplot as plt
import numpy as np
df_new = df.groupby('model_year').count()[['name']]
df_new.reset_index(inplace = True)
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
#plotting the lines
ax.bar(df_new['model_year'],df_new['name'])
```

```python
import matplotlib.pyplot as plt
import numpy as np
new_df = df.groupby('origin').count()[['name']]
new_df.reset_index(inplace=True)
colors = ['red', 'green', 'blue' ]
explode = [0.05, 0, 0]
#creating an empty canvas/figure
fig = plt.figure()
#setting axes
ax = fig.add_axes([0,0,1,1])
#plotting the lines
ax.pie(new_df['name'], labels=new_df['origin'], colors = colors, explode = explode, autopct='%0.1f%%')
```
