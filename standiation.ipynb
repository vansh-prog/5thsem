```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn .preprocessing import StandardScaler
```

```python
df = pd.read_csv('dataset/auto_mpg.csv',usecols=[0,3])
df
```

```python
sns.scatterplot(data=df, x='mpg', y='horsepower')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(df)
df_scaled = scaler.transform(df)
```

```python
df_scaled
```

```python
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled
```

```python
sns.scatterplot(data=df_scaled, x=df_scaled['mpg'], y=df_scaled['horsepower'])
plt.show()
```
