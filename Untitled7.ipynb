```python
import pandas as pd
df = pd.read_csv('dataset/auto_mpg.csv')
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(data=df, y='mpg')
plt.show()
```

```python
sns.scatterplot(data=df,x='horsepower', y='mpg')
plt.show()
```

```python
sns.countplot(data=df,x='model_year')
plt.show()
```

```python
sns.histplot(data=df,x='mpg',kde=True,bins=20)
plt.show()
```

```python
sns.barplot(data=df,x='origin',y='model_year')
```

```python
sns.barplot(data=df,y='mpg',x='model_year')
```

```python
sns.pairplot(data=df)
```

```python
sns.heatmap(data=df[['mpg','model_year','cylinders']])
```
