```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('dataset/SaleData.xlsx')
```

```python
df
```

```python
df.dropna()
```

```python
df1 = df.groupby(df['Region']).sum()['Sale_amt']
df1
```

```python
import matplotlib.pyplot as plt
df1 = df.groupby(df['Region']).sum()['Sale_amt']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Total sale amount as region')
ax.set_xlabel('Region')
ax.set_ylabel('Total sale_amt')
ax.bar(df1.index,df1)
```

```python
import matplotlib.pyplot as plt
df1 = df.groupby(df['Item']).sum()['Units']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Items sold pre units')
ax.set_xlabel('Item')
ax.set_ylabel('Units')
ax.bar(df1.index,df1)
```

```python
df1 = df.groupby(df['Item']).sum()['Units']
df1
```

```python
import matplotlib.pyplot as plt
df1 = df.groupby(df['Manager']).sum()['Sale_amt']
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Sale amount by manager')
ax.set_xlabel('Manager')
ax.set_ylabel('sale_amt')
ax.bar(df1.index,df1)
```
